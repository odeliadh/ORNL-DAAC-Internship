## Overview 

This tutorial will demonstrate Earthdata manipulation and visualization. We'll explore archived airborne data using Earthdata Search and then, through a Notebook Tutorial, programmatically access, visualize, and manipulate Level 3 AVIRIS-Next Generation (ANG) Reflectance dataset collected during the Biodiversity Survey of the Cape (BioSCape) Campaign. We will also visualize additional BioSCape data. 

AVIRIS is an acronym for Airborne Visible InfraRed [**Image Spectrometer**](https://www.researchgate.net/profile/Alexander-Goetz-2/publication/24363398_Terrestrial-Imaging_Spectroscopy/links/544a6f800cf2f2f6012ada0d/Terrestrial-Imaging-Spectroscopy.pdf). Image spectroscopy gathers both spatial and spectral information about a scene. AVIRIS-NG uses hyperspectral image spectroscopy, which collects spectral data in many narrow, contiguous spectral bands. This kind of data is often referred to as a data cube, wherein the X and Y axes represent the spatial coordinates, and the Z axis represents wavelength. 

AVIRIS-NG operates within the 380-2510nm range with ~5nm sampling. AVIRIS-NG data consists of 425 contiguous spectral bands.

![Data Cube](../images/Hyperspectral_image_data_cube.jpg "Credit: University of Texas at Austin, Center for Space Research")

- Learn more about [**AVIRIS instruments**](https://aviris.jpl.nasa.gov/)
- Learn more about [**AVIRIS-Next Generation**](https://avirisng.jpl.nasa.gov/)

BioSCape is an integrated field and airborne campaign in South Africa's Greater Cape Floristic Region (GCFR) where collections occurred in 2023.  The BioSCape Campaign utilized four NASA airborne instruments to collect UV/visible to short-wavelength infrared (UVSWIR) and thermal imaging (TIR) spectroscopy and laser altimetry LiDAR data over terrestrial and aquatic targets.   AVIRIS-NG, Portable Remote Imaging SpectroMeter (PRISM), Land, Vegetation, and Ice Sensor (LVIS), and Hyperspectral Thermal Emission Spectrometer (HyTES).

- Learn more about [**BioSCape**](https://www.bioscape.io/)
- Nasa Earthdata Search: [**Discover Earthdata BioSCape Data**](https://search.earthdata.nasa.gov/search?fpj=BioSCape)

The BioSCape Campaign has produced an AVIRIS-NG L3 Resampled Mosaic dataset.  
-  Surface reflectance data (Level 2b) derived from the AVIRIS-NG instrument were resampled to 5-m spatial resolution and mosaiced into a regular tile system of 807 tiles. A given tile includes multiple AVIRIS-NG scenes from multiple flight lines spanning multiple days.  The mosaics in this dataset were generated by stitching together separate scenes and resampling to 5-m resolution in the Hartebeesthoek94 projected coordinate system (WGS-84 datum, EPSG 9221). The mosaic files are distributed on a tiled grid system, and the tile name is included in the file name. Mosaics were generated by manually grouping sets of flight lines into different chunks that should be placed 'in front of' or 'behind' other chunks. The selection criteria included a combination of the weather during observations, flight conditions, flightbox design, and the time the flight was taken. 

> Brodrick, P.G., A.M. Chlus, R. Eckert, J.W. Chapman, M. Eastwood, S. Geier, M. Helmlinger, S.R. Lundeen, W. Olson-Duvall, R. Pavlick, L.M. Rios, D.R. Thompson, and R.O. Green. 2025. BioSCape: AVIRIS-NG L3 Resampled Reflectance Mosaics, V2. ORNL DAAC, Oak Ridge, Tennessee, USA. [https://doi.org/10.3334/ORNLDAAC/2427](https://doi.org/10.3334/ORNLDAAC/2427)

Dataset [**Data Processing Levels**](https://www.earthdata.nasa.gov/learn/earth-observation-data-basics/data-processing-levels) 
- **Level 3**:  Variables mapped on uniform space-time grid scales, usually with some completeness and consistency.

## Step 2: Further Define and Refine Search Parameters
### 2.1 Create and Visualize a **`bounding box`** for an Area-of-Interest within the South Africa Greater Cape Floristic Region (GCFR) of the BioSCape Campaign 

```python
def create_geo_bb(coordinates, crs_in='epsg:4326', crs_out='epsg:4326'):
    
    polygon_shape = Polygon(coordinates)

    if crs_in != crs_out:
      project_in = pyproj.Proj(init=crs_in)
      project_out = pyproj.Proj(init=crs_out)
    
      polygon_shape = transform(pyproj.Transformer.from_proj(project_in, project_out, always_xy=True).transform, polygon_shape)
    
    polygon = gpd.GeoDataFrame(geometry=[polygon_shape], crs=crs_out)
    return polygon

# Define the area you want to focus on
coordinates = [
    (18.5144, -34.0774),
    (18.5144, -34.3727),
    (18.3300, -34.3727),
    (18.3300, -34.0774),
    (18.5144, -34.0774)
]

polygon_gdf = create_geo_bb(coordinates)
print(polygon_gdf)
```

```python
polygon_gdf.explore(fill=False, tiles='https://mt1.google.com/vt/lyrs=s&x={x}&y={y}&z={z}', attr='Google')
```

**We'll use this bounding box and temporal parameters to refine our search of BioSCape AVIRIS-NG L3 files in our region and time of interest**

Again, using **`earthaccess`** we can query the [**`BioSCape: AVIRIS-NG L3 Resampled Reflectance Mosaics, V2`**](https://search.earthdata.nasa.gov/search/granules?p=C3523930138-ORNL_CLOUD) data to discover files within the spatial and temporal subset of interest.

```python
# bounding lon, lat as a list of tuples
bounds = polygon_gdf.geometry.apply(orient, args=(1,))
# simplifying the polygon to bypass the coordinates 
# limit of the CMR with a tolerance of .01 degrees
xy = bounds.simplify(0.01).get_coordinates()

date_range = ("2023-10-19", "2023-11-30")  

results = earthaccess.search_data(
    short_name = 'BioSCape_ANG_V02_L3_RFL_Mosaic_2427', 
    polygon=list(zip(xy.x, xy.y)),
    temporal = date_range,
    granule_name=('*AVIRIS-NG_BIOSCAPE_V02_L3*')
)
print(f"Total granules found: {len(results)}")
```

### 2.2 Create and Visualize the **`Bounding Boxes`** of the subset of files
From each granule, we'll use the CMR `Geometry` information to create a plot of the AVIRIS-3 flight lines from our temporal and spatial subset

Below, we define two functions to plot the search results over a basemap
- Function 1:  converts UMM geometry to multipolygons
  -- `UMM` stands for NASA's Unified Metadata Model
- Function 2: converts the Polygon List [ ] to a **`geopandas`** dataframe

```python
def convert_umm_geometry(gpoly):
    """converts UMM geometry to multipolygons"""
    multipolygons = []
    for gl in gpoly:
        ltln = gl["Boundary"]["Points"]
        points = [(p["Longitude"], p["Latitude"]) for p in ltln]
        multipolygons.append(Polygon(points))
    return MultiPolygon(multipolygons)

def convert_list_gdf(datag):
    """converts List[] to geopandas dataframe"""
    # create pandas dataframe from json
    df = pd.json_normalize([vars(granule)['render_dict'] for granule in datag])
    # keep only last string of the column names
    df.columns=df.columns.str.split('.').str[-1]
    # convert polygons to multipolygonal geometry
    df["geometry"] = df["GPolygons"].apply(convert_umm_geometry)
    # return geopandas dataframe
    return gpd.GeoDataFrame(df, geometry="geometry", crs="EPSG:4326")
    
subset_gdf = convert_list_gdf(results)
subset_gdf.crs
#subset_gdf.drop('Version', axis=1, inplace=True)
#subset_gdf.explore(fill=False, tiles='https://mt1.google.com/vt/lyrs=s&x={x}&y={y}&z={z}', attr='Google')
```

```python
# let's visualize the bounding boxes of the selected files
subset_gdf = convert_list_gdf(results)

mapObj = folium.Map(location=[-34.25, 18.4], zoom_start=10, control_scale=True)
#-118.2036, 34.2705
#folium.GeoJson(gdf).add_to(mapObj)
subset_gdf.drop('Version', axis=1, inplace=True)
folium.GeoJson(subset_gdf, name="SUBSET FLIGHT LINES", color="blue", style_function=lambda x: {"fillOpacity": 0}).add_to(mapObj)

folium.GeoJson(polygon_gdf, name="PENINSULA SUBSET AREA", color="white", style_function=lambda x: {"fillOpacity": 0}).add_to(mapObj)
#folium.GeoJson(lasubset, name="SUBSET FLIGHT LINES", style_function=lambda x: {"fillOpacity": 0}).add_to(mapObj)

# create ESRI satellite base map
esri = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}'
folium.TileLayer(tiles = esri, attr = 'Esri', name = 'Esri Satellite', overlay = False, control = True).add_to(mapObj)
folium.LayerControl().add_to(mapObj)
mapObj
```

### 4.2 Visualize our [**`BioSCape: AVIRIS-NG L3 Resampled Reflectance Mosaics, V2`**](https://search.earthdata.nasa.gov/search/granules?p=C3523930138-ORNL_CLOUD) subset **`bounding boxes`** over our **`vegetation centers`** file

```python
file_objs = RFLnc_paths

target_nc = "AVIRIS-NG_BIOSCAPE_V02_L3_38_11_RFL.nc"

# Get the in-memory file-like object
file_idx = [i for i, path in enumerate(nc_r) if target_nc in path][0]
f = file_objs[file_idx]

# Create plot
fig, ax = plt.subplots(figsize=(10, 8))

# Color map for bounding boxes
num_files = len(nc_r)
colors = plt.colormaps['tab10'].resampled(num_files)
legend_handles = []

for i, (path, f) in enumerate(zip(nc_r, file_objs)):
    # Open the .nc file
    ds = xr.open_dataset(f, engine="h5netcdf")
    east = ds['easting'].values
    north = ds['northing'].values

    # Transform to WGS84
    proj_tm = pyproj.CRS.from_cf(ds["transverse_mercator"].attrs)
    transformer = pyproj.Transformer.from_crs(proj_tm, "EPSG:4326", always_xy=True)
    lon_min, lat_min = transformer.transform(east.min(), north.min())
    lon_max, lat_max = transformer.transform(east.max(), north.max())

    # Create bounding box
    bbox = gpd.GeoDataFrame(
        {"filename": [path.split("/")[-1]]},
        geometry=[box(lon_min, lat_min, lon_max, lat_max)],
        crs="EPSG:4326"
    )

    color = colors(i)
    bbox.plot(ax=ax, edgecolor=color, facecolor='none', linewidth=2)

    # Add legend handle
    patch = mpatches.Patch(edgecolor=color, facecolor='none', label=path.split("/")[-1], linewidth=2)
    legend_handles.append(patch)

# Plot shapefile
clipped_gdf.plot(ax=ax, facecolor='none', edgecolor='black', linewidth=1)

# Add shapefile legend entry
legend_handles.append(mpatches.Patch(edgecolor='black', facecolor='none', label='Shapefile', linewidth=1))

# Refine legend specifications
plt.legend(
    handles=legend_handles,
    loc="center left",
    bbox_to_anchor=(1, 0.5),  # (x=1 is to the right of the axes)
    fontsize="small",
    borderaxespad=0
)

plt.title("Overlay: Bounding Boxes + Shapefile")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
transformer = pyproj.Transformer.from_crs("EPSG:4326", "EPSG:9221", always_xy=True)

spectra = []

for i, row in clipped_gdf.iterrows():
    lon, lat = row.geometry.x, row.geometry.y
    easting, northing = transformer.transform(lon, lat)

    x_idx = np.abs(rfl_ds["easting"].values - easting).argmin()
    y_idx = np.abs(rfl_ds["northing"].values - northing).argmin()
    
    # Step 1: Get the reflectance spectrum from one pixel
    rfl_array = rfl_ds["/reflectance"].data_vars["reflectance"]
    spectrum = rfl_array.isel(easting=x_idx, northing=y_idx)

    # Step 2: Convert bblist to a DataArray with matching wavelength coordinates
    wavelengths = spectrum["wavelength"]
    bbl_mask = xr.DataArray((bblist != 0), dims=["wavelength"], coords={"wavelength": wavelengths})

    # Step 3: Optionally restrict to physical range
    range_mask = (wavelengths > 400) & (wavelengths < 2400)

    # Step 4: Combine masks and apply
    mask = bbl_mask & range_mask
    spectrum = spectrum.where(mask)

    spectra.append(spectrum)
    
plt.rcParams['figure.figsize'] = [10,7]

for i, spectrum in enumerate(spectra):
    spectrum.plot.line(label=plotname_plant[i], alpha=0.7)

plt.xlabel('AVIRIS-NG wavelength', fontsize=12)
plt.ylabel('reflectance', fontsize=12)
plt.title("Reflectance spectra at selected points")
plt.legend(
    loc='center left',
    bbox_to_anchor=(1.02, 0.5),
    fontsize="small",
    borderaxespad=0,
    title="Genus_species"
)
plt.show()
```

## Step 6: Visualize Spectral Profiles for a Plot using a 3x3 Pixel Average Surrounding the Plot

**The Histogram we just visualized looks at the wavelengths of an individual pixel where a given plot lies. If we take an average of the pixels surrounding the plot, we can have a better idea of what the spectral profiles look like for our dataset.**

### 6.1 Extract the first three concatenated 3x3 pixel plots from **`clipped_gdf`**

```python
# Load reflectance array once
rfl_array = rfl_ds["/reflectance"].data_vars["reflectance"]

# Get wavelength vector and good band indices
wavelength = rfl_ds["/reflectance"].coords["wavelength"].values
good_idx = np.where((wavelength > 400) & (wavelength < 2400))[0]
wavelength = wavelength[good_idx]

# Define a function to get a 3x3 block of filtered reflectance
def extract_filtered_cube(lon, lat):
    try:
        # Convert from geographic (lon/lat) to AVIRIS projected coordinates
        easting, northing = transformer.transform(lon, lat)
        x_idx = np.abs(rfl_ds["easting"].values - easting).argmin()
        y_idx = np.abs(rfl_ds["northing"].values - northing).argmin()
        x_slice = slice(max(0, x_idx - 1), x_idx + 2)
        y_slice = slice(max(0, y_idx - 1), y_idx + 2)

        # Pull the 33bands reflectance block as a DataArray
        block = rfl_array.isel(easting=x_slice, northing=y_slice)
        block = block.transpose("easting", "northing", "wavelength")
        block = block.where(bblist != 0)
        refl_block = block.values.reshape(-1, block.shape[-1])
        refl_block = refl_block[:, good_idx] # Mask "bad bands"

        # Validate the result
        valid_mask = ~np.all(np.isnan(refl_block), axis=1)
        if not np.any(valid_mask):
            return None

        return refl_block[valid_mask]

    except Exception as e:
        print(f" Failed extraction at lon={lon}, lat={lat}: {e}")
        return None
```

```python
dfs = []
for i, (row, label) in enumerate(zip(clipped_gdf.iterrows(), plotname_plant)):
    # Pull lon and lat from plot point geometry, extract a filtered and cleaned 33 reflectance block
    lon, lat = row[1].geometry.x, row[1].geometry.y
    refl_block = extract_filtered_cube(lon, lat)
    
    if refl_block is None:
        print(f"Skipped Plot {row[1]['BScpPID']} due to no valid data")
        continue

    # Create a mini-DataFrame for this plot with plot, name, and reflectance
    df = pd.DataFrame({
        "Plot": [row[1]["BScpPID"]],
        "Name": [label],
        "reflectance": [refl_block]
    })
    dfs.append(df)

# Combine all the one-row DataFrames into a full table with plot, name, and reflectance (one row per valid plot)
avng_rfl_df = pd.concat(dfs, ignore_index=True)
final_gdf = pd.merge(filtered_veg, avng_rfl_df, on="Plot", how="left")
```

### 6.2 Visualize the spectral profiles of the first three 3x3 pixel plots including their Mean & Standard Deviation

```python
geo_df = clipped_gdf.rename(columns={"BScpPID": "Plot"})[["Plot", "geometry"]]
# Merge reflectance + species data (final_gdf) with geometry data (geo_df) using Plot as the key
final_gdf_with_geom = pd.merge(final_gdf, geo_df, on="Plot", how="left")
# Convert the merged table into a proper GeoDataFrame
final_gdf_with_geom = gpd.GeoDataFrame(final_gdf_with_geom, geometry="geometry", crs=clipped_gdf.crs)

# Select first three valid plots with geometry and only one row per plot
first_three = final_gdf_with_geom.dropna(subset=["geometry"]).drop_duplicates("Plot").head(3).copy()
first_three["lon"] = first_three.geometry.x
first_three["lat"] = first_three.geometry.y

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

for i, (_, row) in enumerate(first_three.iterrows()):
    refl_block = extract_filtered_cube(row["lon"], row["lat"])
    if refl_block is None:
        print(f" No valid reflectance for Plot {row['Plot']}")
        continue

    # Compute the mean and standard deviation across all valid pixels in the 33 block
    mean_reflectance = np.nanmean(refl_block, axis=0)
    std_reflectance = np.nanstd(refl_block, axis=0)

    # Define plot display
    ax = axes[i]
    ax.plot(wavelength, mean_reflectance, label="Mean Reflectance")
    ax.fill_between(wavelength,
                    mean_reflectance - std_reflectance,
                    mean_reflectance + std_reflectance,
                    alpha=0.3,
                    label="1 STD")
    ax.set_title(f"Plot {row['Plot']}")
    ax.set_xlabel("Wavelength (nm)")
    if i == 0:
        ax.set_ylabel("Reflectance")
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()
```

## Step 7: Visualize Normalized Difference Vegetation Index (NDVI) for AVIRIS data

**When dealing with hyperspectral data, we have countless spectral indices we could calculate. NDVI is one of those spectral indicies, calculated here.** 

### 7.1 Select Data to Perform NDVI on

```python
ds = xr.open_dataset(RFLnc_paths[5].f, group="/reflectance")

# Extract the actual reflectance DataArray from the dataset
rfl = ds["reflectance"]

print(rfl.dims)

# Grab the wavelength coordinate as a NumPy array
wavelengths = (rfl["wavelength"].values)
print(wavelengths[(wavelengths >= 650) & (wavelengths <= 670)])
print(wavelengths[(wavelengths >= 810) & (wavelengths <= 830)])
```

### 7.2 Calculate and Visualize NDVI

```python
# Select red & nir bands from list in previous output
red = rfl.sel(wavelength=657.68567)
nir = rfl.sel(wavelength=822.96564)

# calculate ndvi
ndvi = (nir - red) / (nir + red)

ndvi.plot(cmap="RdYlGn")
plt.title("NDVI")
# Invert the y-axis so it matches north-up geospatial convention
plt.gca().invert_yaxis()
plt.show()
```
